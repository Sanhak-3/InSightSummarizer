```python
#https://huggingface.co/ainize/kobart-news
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

# 모델 이름 목록
model_names = [
    #"ainize/kobart-news"
    #"noahkim/KoT5_news_summarization"
    'KETI-AIR-Downstream/long-ke-t5-base-summarization'
    #KETI-AIR/ke-t4-base 는 기본 모델
]

# 뉴스 요약 함수
def summarize_news(model_name, news_text):
    try:
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        #tokenizer.pre_tokenizer = Whitespace() #error : data did not match any variant of untagged enum PyPreTokenizerTypeWrapper at line 1 column 26564
        model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
        inputs = tokenizer(news_text, return_tensors="pt", max_length=1024, truncation=True, padding="max_length")
        summary_ids = model.generate(inputs['input_ids'], max_length=150, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        return summary
    except Exception as e:
        return f"오류 발생: {e}"

# 뉴스 요약 및 출력 함수
def summarize_and_display_news(news_text):
    print(f"뉴스 원문:\n{news_text}\n")
    print("요약 결과:\n")
    for model_name in model_names:
        model_summary = summarize_news(model_name, news_text)
        print(f"모델: {model_name}")
        print(f"요약문: {model_summary}\n")
        print('-' * 80)
```

```python
# 뉴스 원문 입력 받기 (Example)
news_text = input("뉴스 원문을 입력하세요: ")

# 뉴스 요약 및 출력
summarize_and_display_news(news_text)
```

```python
df1['ours'] = '' #초기화
df2['ours'] = ''
df3['ours'] = ''
for i in range(3):
    for j in range(10):
        df1['ours'][j] = summarize_news('KETI-AIR-Downstream/long-ke-t5-base-summarization', df1['document'][j])
        df2['ours'][j] = summarize_news('KETI-AIR-Downstream/long-ke-t5-base-summarization', df2['document'][j])
        df3['ours'][j] = summarize_news('KETI-AIR-Downstream/long-ke-t5-base-summarization', df3['document'][j])
```
