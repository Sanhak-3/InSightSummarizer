```python
import pandas as pd
path = '/content/gdrive/MyDrive/Colab Notebooks/LLM평가/evaluated'
df11 = pd.read_csv(path+'1_1th.csv')
df21 = pd.read_csv(path+'2_1th.csv')
df31 = pd.read_csv(path+'3_1th.csv')
df12 = pd.read_csv(path+'1_2th.csv')
df22 = pd.read_csv(path+'2_2th.csv')
df32 = pd.read_csv(path+'3_2th.csv')
```

```python
df1 = (df11+df12)/2
df2 = (df21+df22)/2
df3 = (df31+df32)/2
```

```python
import pandas as pd
#df 칼럼명 바꾸기

df1.columns = [col.replace('_1th', '_avg') for col in df1.columns]
df2.columns = [col.replace('_1th', '_avg') for col in df2.columns]
df3.columns = [col.replace('_1th', '_avg') for col in df3.columns]

df1.head()
```

```python
df1.to_csv(f'/content/gdrive/MyDrive/Colab Notebooks/LLM평가/eval_avg1.csv', index=False,encoding='utf-8-sig')
df2.to_csv(f'/content/gdrive/MyDrive/Colab Notebooks/LLM평가/eval_avg2.csv', index=False,encoding='utf-8-sig')
df3.to_csv(f'/content/gdrive/MyDrive/Colab Notebooks/LLM평가/eval_avg3.csv', index=False,encoding='utf-8-sig')
```

```python
import matplotlib.pyplot as plt

#세번 복원추출한 거 평균, 10번의 평균
df = (df1+df2+df3)
# Rename columns
df.columns = [col.replace('_1th', '_avg') for col in df.columns]
df.to_csv(f'/content/gdrive/MyDrive/Colab Notebooks/LLM평가/eval_avg_last.csv', index=False,encoding='utf-8-sig')
df = df.mean(axis=0)

coherence = df[['coherence_summary_avg','coherence_ChatGPT4o_avg','coherence_ChatGPT35_avg',
                'coherence_Claude_avg', 'coherence_Gemini_avg','coherence_ours_avg']]
faithfullness = df[['faithfullness_summary_avg','faithfullness_ChatGPT4o_avg','faithfullness_ChatGPT35_avg',
                'faithfullness_Claude_avg', 'faithfullness_Gemini_avg','faithfullness_ours_avg']]
toxicity = df[['toxicity_summary_avg','toxicity_ChatGPT4o_avg','toxicity_ChatGPT35_avg',
                'toxicity_Claude_avg', 'toxicity_Gemini_avg','toxicity_ours_avg']]
summarization = df[['summarization_summary_avg','summarization_ChatGPT4o_avg','summarization_ChatGPT35_avg',
                'summarization_Claude_avg', 'summarization_Gemini_avg','summarization_ours_avg']]

plt.figure(figsize=(10, 8))
plt.subplot(221)
plt.bar(['Summary','ChatGPT4o','ChatGPT3.5','Claude','Gemini','Ours'], coherence.values,
        color=['#1f77b4','#1f77b4','#1f77b4','#1f77b4','#1f77b4','#FF7F0F'])

#for i in range(6): #숫자표시
#  plt.text(i, coherence.values[i]+0.01, str(round(coherence.values[i],1)), ha='center')
plt.title('Coherence')
plt.ylabel('Average Score')
plt.xticks(rotation=45)

plt.subplot(222)
plt.bar(['Summary','ChatGPT4o','ChatGPT3.5','Claude','Gemini','Ours'], faithfullness.values,
        color=['#1f77b4','#1f77b4','#1f77b4','#1f77b4','#1f77b4','#FF7F0F'])
#for i in range(6):
#  plt.text(i, faithfullness.values[i]+0.01, str(round(faithfullness.values[i],1)), ha='center')
plt.title('Faithfulness')
plt.ylabel('Average Score')
plt.xticks(rotation=45)

plt.subplot(223)
plt.bar(['Summary','ChatGPT4o','ChatGPT3.5','Claude','Gemini','Ours'], toxicity.values,
        color=['#1f77b4','#1f77b4','#1f77b4','#1f77b4','#1f77b4','#FF7F0F'])
#for i in range(6):
#  plt.text(i, toxicity.values[i]+0.01, str(round(toxicity.values[i],1)), ha='center')
plt.title('Toxicity')
plt.ylabel('Average Score')
plt.xticks(rotation=45)

plt.subplot(224)
plt.bar(['Summary','ChatGPT4o','ChatGPT3.5','Claude','Gemini','Ours'], summarization.values,
        color=['#1f77b4','#1f77b4','#1f77b4','#1f77b4','#1f77b4','#FF7F0F'])
#for i in range(6):
#  plt.text(i, summarization.values[i]+0.01, str(round(summarization.values[i],1)), ha='center')
plt.title('Summarization')
plt.ylabel('Average Score')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
